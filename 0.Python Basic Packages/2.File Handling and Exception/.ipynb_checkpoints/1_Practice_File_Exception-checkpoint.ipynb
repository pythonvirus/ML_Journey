### operation on CSV File

```python
csvfile = open(r'C:\Users\Ankush\Desktop\Python\Meetup\FileHandling\Zensar.csv','r')
filelines = csvfile.readlines()
mydict = dict()
```

```python
cntr = 0
for line in filelines:
    col = line.split(',')
    cntr += 1
    #print(cntr, ' ', col)
    year = col[0].split('-')[0]
    price = col[4]
    
    if col[0] == 'Date':
        continue
    
    #print('({}) year = {} \t price = {}'.format(cntr,year,price))
    
    if year not in mydict:
        pricelist = list()
        pricelist.append(float(price))
        mydict[year] = pricelist
    else:
        mydict[year].append(float(price))
        
print(mydict)
        
    
```

```python
myfile2 = open(r'C:\Users\Ankush\Desktop\Python\Meetup\FileHandling\CSVAnswer.txt','w')
#for year,price in sorted(mydict.items()):
for year,price in mydict.items():
    #print('Year {} has avg price {}' .format(year, sum(price)/len(price)))
    myfile2.write('Year {} has avg price {}' .format(year, round(sum(price)/len(price),2))+'\n')


myfile2.write('\n\n')
#myfile2.write(str(mydict))
#print(mydict)

myfile2.close()
```

### sys.exc_info() Function

- It gives us the Exception type

```python
import sys
try:
    print("In try...")
    #a = 10/0
    str = "Ankush"
    print(str[10])
    
except:
    print("In except...")
    print("Oops! Exception Occur : ", sys.exc_info())
    print("Length : ", len(sys.exc_info()))
    print("Length : ", type(sys.exc_info()))
    print("sys[0] :", sys.exc_info()[0])
    #print("sys[1] :", sys.exc_info()[1])
    #print("sys[2] :", sys.exc_info()[2])
    
else:
    print("In else...")
finally:
    print("In finally...")
```

### Argument of an Exception

### raise keyword

```python
import sys
try:
    a = int(input("Enter 1st value... "))
    b = int(input("Enter 2nd value... "))

    if(b <= 0):
        raise ZeroDivisionError ("Number is not a positive...")
    else:
        c = a/b

except ZeroDivisionError as err:
    print("Oops! Exception :", err)

except:
    print("In except...")
    print("Oops! Exception Occur : ", sys.exc_info())
    
else:
    print("Division is :", c)
    
finally:
    print("In finally...")
```

### Assertion

- Assertions are simply boolean expressions that checks if the conditions return true or not. If it is true, the program does nothing and move to the next line of code. However, if it's false, the program stops and throws an error. 
- Basically use for debugging.
- Syntex --> **assert CONDITION, "MESSAGE"**

```python
num=int(input('Enter a number: '))
assert num >= 0, "Number should be greater or equal to zero."
print("Number is :", num)
```

```python
def KelToFehr(Temprature):
    assert (Temprature >= 0), "Colder than absolute Zero"
    return ((Temprature-273)*1.8)+32

print("Temprature in Fehrhenite :", KelToFehr(int(input("Enter the Temprature in Kelvin... "))))
```

```python
try:
    #mylst = []
    mylst = [10,20,30,40,50]
    assert len(mylst) != 0, "List is Empty."
    print("List is ", mylst)
    
except AssertionError as err:
    print("Catch exception! : ", err)
except:
    print("Other Exception...")
    
else:
    print("Avrage of list is ", sum(mylst)/len(mylst))
    
finally:
    print("Program End")
```
